```python
def visualisator(data):
    # визуализируем распределение
    print('\033[1m' + 'РАСПРЕДЕЛЕНИЕ ПРИЗНАКОВ' + '\033[0m')
    data.hist(
        bins=50, figsize=(20,15), 
        color='r', edgecolor='black', 
        linewidth=1
    )
    plt.show()
    print()
    
    # нарисуем heatmap
    plt.figure(figsize = (15,15))
    plt.title(f'Корреляция признаков', fontsize=18)
    sns.heatmap(
        data.corr(), square=True, 
        annot=True, linewidths=1, 
        fmt='.3f', linecolor='gray',
        vmin=-1, vmax=1,
        cmap='seismic'
    )
    plt.show()    
```

### Изучение выбросов

```python
def boxplot_slave(data, name, xlim = None, ylim = None):
    '''
    boxplot_slave принимает данные в параметр data и возвращает боксплот;
    name используется для разметки и принимает название исследуемого признака.
    xlim и ylim используются для масштабирования графика, ограничивая его оси
    и принимают кортеж (tuple) из двух цифр (начало и конец ограничения)
    '''
    # выведем описание данных
    print('\033[1m' + f'Описание признака {name}:' + '\033[0m')
    print(data.describe())
    print('=' * 45)
    
    # построим боксплот
    box = plt.boxplot(
        x=data, vert=False, 
        patch_artist=True, notch=True
    )
    fig = plt.gcf()
    fig.set_size_inches(15, 8)
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
    plt.title(
            f'Боксплот признака {name}', 
            fontsize=14
        )
    plt.show()
    
    # выведем его минимальные и максимальные значения
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - (1.5 * IQR)
    upper_fence = Q3 + (1.5 * IQR)
    print('\033[1m' + f'Пороговые значения боксплота {name}:' + '\033[0m')
    print('Нижняя граница значений:', lower_fence)
    print('Верхняя граница значений:', upper_fence)
    print()
    print()
```

## Статистический анализ

По результатам проведенного исследования возникает вопрос о статистическом различии представленных кластеров. Из всех кластеров 3 имеют крайне схожие значения и только 1 (имеющий 56 наблюдений) отличается. При этом у кластеров различный параметр оттока. Можно сделать предположение, что наличествует техническая ошибка в сборе данных. Либо же ключевым значением является баланс, который и оказывает формирование на поведение клиента.

Для проверки, сперва расчитаем средние значения и дисперсии для каждого кластера.

После чего проверим сравниваемые выборки на гомогенность дисперсий. Для чего используем тест Левена. 

При этом, если p-уровень значимости будет меньше критического уровня статистической значимости (т.е. вероятность равенства дисперсий достаточно низкая), то при проверке нулевой гипотезы необходимо использовать t-критерий Велча (использовать параметр equal_var = False), иначе будем использовать equal_var = True, для выборок с высокой вероятностью равенства дисперсий.

Определим необходимые функции.

```python
# функция проводит статистические тесты
def stat_slave(data_1, data_2, feature, alpha=0.05):
    '''
    Функция stat_slave проводит статистический тест и принимает следующие параметры:
    data_1, data_2 - наборы данных для проведения теста;
    alpha - критический уровень статистической значимости.
    '''

    # проведение теста Левена
    results_var_test = st.levene(data_1[feature], data_2[feature], center='mean')

    # вывод на экран 
    print('p-значение для теста Левена: {:.2%}'.format(results_var_test.pvalue))

    # сравнение p-значение
    if (results_var_test.pvalue < alpha):     
        print('Отвергаем гипотезу о равенстве дисперсий сравниваемых выборок.')
        equal_var_test = True
    else:
        print('Не получилось отвергнуть гипотезу о равенстве дисперсий сравниваемых выборок.')
        equal_var_test = False
        
    # проведение статистического теста
    results = st.ttest_ind(
        data_1[feature], data_2[feature],
        equal_var = equal_var_test)

    # вывод на экран
    print('p-значение для t-теста: {:.2%}'.format(results.pvalue))

    # сравним получившееся p-значение
    if (results.pvalue < alpha):
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

```python
# функция расчитывает среднее и дисперсии выбранного признака
def math_slave(data_1, data_2, feature):
    '''
    Функция math_slave расчитывает среднее и дисперсию переданных data_1 и data_2.
    '''
    # расчеты
    mean_1 = data_1[feature].mean()
    var_1 = data_1[feature].var()
    mean_2 = data_2[feature].mean()
    var_2 = data_2[feature].var()
    
    # выводы
    print('Cредний доход клиентов, использующих {} продукта: {:.1f}'.format(set(data_1['products']), mean_1))
    print('Дисперсия доходов клиентов, использующих {} продукта: {:.1f}'.format(set(data_1['products']), var_1))
    print('Cредний доход клиентов, использующих {} продукта: {:.1f}'.format(set(data_2['products']), mean_2))
    print('Дисперсия доходов клиентов, использующих {} продукта: {:.1f}'.format(set(data_2['products']), var_2))
```
