Подавляющее большинство значений тяготеет к резонным значениям. Значения свыше массовых встречаются редко. Есть значительное искажение в области высоты потолка 20+ метров. Согласно гистограмме складывается ощущение, что произошла ошибка ввода разделителя высоты потолков (вместо 2.7 внесено 27.0). Высота потолка в 100 метров маловероятна для помещения, используемого в качестве апартаментов, общей площадью 25 кв.м. В столбце определенно есть ошибки. При этом, высота потолка является косвенным признаком класса жилья (более дорогое жилье тяготеет к более высоким потолкам), поэтому полностью исключать этот показатель из исследования нельзя. Однако, с определенного момента, высота потолка уже не будет иметь значение. Определим разумные границы значений от 2.4 (нормативная высота потолков при строительстве в РФ) до 5.0 метров (максимальная высота потолков не устанавливается нормативными документами, но в старом жилом фонде может достигать значительных значений), а все, что выходит за эти пределы, приравняем к этим значениям.

```python
# напишем две функции для замены значений. 
# Не будем объединять их, в одну, так как в дальнейшем  
# они могут нам понадобиться раздельно для работы со столбцами

# функция делитель

def divide_by(value_for_change, value_min, value_max, divider):
    '''
    Функция divide_by сравнивает значение value_for_change,
    с фильтрами value_min и value_max, и, если значение value_for_change
    находится между значениями фильтров, то возвращает значение value_for_change, 
    разделенное на значение, указанное в параметре divider.
    Если же значение value_for_change находится за пределами фильтров,
    то функция возвращает значение value_for_change без изменений.
    '''
    if value_min < value_for_change < value_max:
        return value_for_change / divider
    return value_for_change

# протестируем функцию

print('Результаты проверки функции divide_by')
print(divide_by(21, 18, 40, 10))
print(divide_by(18, 18, 40, 10))
print(divide_by(36, 18, 40, 10))
print(divide_by(42, 18, 40, 10))
print()

# функция заменитель

def value_changer(value_for_change, value_min, value_max, new_value):
    '''
    Функция value_changer сравнивает значение value_for_change,
    с фильтрами value_min и value_max, и возвращает фиксированное значение, 
    указанное в параметре new_value, если значение value_for_change
    меньше value_min или больше value_max. Если же значение value_for_change
    находится между фильтрами, то функция возвращает 
    значение value_for_change без изменений.
    '''
    if value_for_change < value_min:
        return new_value
    if value_for_change > value_max:
        return new_value
    return value_for_change

# протестируем функцию

print('Результаты проверки функции value_changer')
print(value_changer(1.9, 2.4, 5.0, 3))
print(value_changer(2.4, 2.4, 5.0, 3))
print(value_changer(4, 2.4, 5.0, 3))
print(value_changer(5.5, 2.4, 5.0, 3))
print(value_changer(100, 2.4, 5.0, 3))
```

```python
# функции работают, произведем замену знаений 'ceiling_height'

df['ceiling_height'] = df['ceiling_height'].apply(
    divide_by, value_min=18, value_max=40, divider=10
)

# после преобразования функцией divide_by два значения подпадут
# под преобразование функцией value_changer, это осознанно,
# так как эти значения меньше нормативных значений высоты потолков

df['ceiling_height'] = df['ceiling_height'].apply(
    value_changer, value_min=2.4, value_max=4.5, 
    new_value=df['ceiling_height'].median()
)

# после изменения данных заполним пропуски

df['ceiling_height'] = df['ceiling_height'].fillna(value=df['ceiling_height'].median())

# проверим изменение данных

print('Количество пропусков в столбце \'ceiling_height\' до замены:', 
      df_control['ceiling_height'].isna().sum())
print('Количество пропусков в столбце \'ceiling_height\' после замены:', 
      df['ceiling_height'].isna().sum())
print('Медианное значение в столбце \'ceiling_height\' до замены: ', 
      df_control['ceiling_height'].median())
print('Медианное значение в столбце \'ceiling_height\' после замены: ', 
      df['ceiling_height'].median())
print('Медианное значение столбца \'ceiling_height\' изменилось на',
       df['ceiling_height'].median() / df_control['ceiling_height'].median()-1, '%.')
print('Другие значения столбца изменились следующим образом:\n',
      df['ceiling_height'].describe() / df_control['ceiling_height'].describe())

df_control['ceiling_height'].plot.hist(bins=10, ylim=(0, 100), legend=True, label='До изменений')
df['ceiling_height'].plot.hist(bins=10, ylim=(0, 100), legend=True, label='После изменений')
plt.show()
```

Проведенная обработка пропусков изменила репрезентативность колонки 'ceiling_height', убрав пиковые значения, которые можно смело считать ошибочными. При этом, медианное значение не изменилось, среднее изменилось всего на 3%, а стандартное отклонение на целых 83%! Учитывая, что данные предоставлены по Санкт-Петербургу, имеющему низовую застройку, а также небольшое количество значений свыше 36 этажей, такую высотность (тем более 52) стоит поставить под вопрос. Кроме того, следует изменить тип данных на int64 (вряд ли можно выделить десятые части этажей). При этом влияние этажности на стоимость объекта будет крайне опосредованным, поэтому пропуски, как и пиковые значения выше 36 этажей, можно смело заполнить медианным значением (повторюсь, архитектурная особенность Санкт-Петербурга подразумевает большое количество малоэтажной застройки).

Для замены значений общего количества этажей необходимо учитывать тот факт, что столбец с номером этажа квартиры заполнен полностью. И необходимо избежать варианта, при котором количество этажей в здании будет меньше этажа квартиры. При замене мы будем ориентироваться на медианное значение, либо на значение этажа квартиры + 1. 


Почему? Это решение выбрано нами на основании гипотезы, что последние этажи имеют тенденцию к небольшому снижению стоимости квартиры. Чтобы избежать этого влияния, мы сделаем количество этажей превышающим этаж квартиры. Кроме того, статистически объектов не на первом/последнем этажах значительно больше, а значит искажение будет минимальным. 


Поэтому, сперва заполним пропуски, а потом обработаем эти пропуски вместе с другими выбивающимися значениями функцией.

```python
# напишем функцию для замены значений

def floor_changer(row):
    '''
    Функция floor_changer принимает в аргументы таблицу, после чего
    сравнивает значение value_for_change с фильтром floor, и возвращает:
    - значение floor + 1, если value_for_change < floor;
    - значение floor + 1, если value_for_change >= floor и value_for_change > 36;
    - значение value_for_change, во всех остальных случаях.
    '''
    value_for_change = row['floors_total']
    floor = row['floor']
    new_value = 9
    if value_for_change < floor:
        return floor + 1
    if value_for_change >= floor and value_for_change > 36:
        return floor + 1
    return value_for_change

df['floors_total'] = df['floors_total'].fillna(value=df['floors_total'].median())
df['floors_total'] = df.apply(floor_changer, axis=1)
df['floors_total'] = df['floors_total'].astype('int')
```

```python
print('Количество пропусков в столбце \'floors_total\' после замены:', 
      df['floors_total'].isna().sum())
print()

print('Итоговая статистика столбца \'floors_total\' после замены:')
print (df['floors_total'].describe())
print()

print('Количество строк, с ошибочным значением total_floor < floor:')
df[df['floors_total'] < df['floor']]
```

По поводу избыточной высотности по прежнему остаются сомнения, однако, небольшое количество, а также гипотеза о малом влиянии количества этажей на цель исследования, вкупе с невозможностью уточнить данные, позволят нам считать столбец подготовленным к исследованиям надлежащим образом.

1. Общая площадь должна минимально превышать жилую во всех объектах.
2. Жилай площадь не должна иметь значения менее 14 кв.м.

Произведем замену ненормативных значений жилых площадей. Согласно технической документации, нормативным соотношением жилой площади к общей является коэффициент 0.5-0.7. Возьмем для учета К=0.6. При этом, этот коэффициент является *рекомендательным*, а не *обязательным*. Кроме того, разумно предположить, что чем больше общая площадь, тем больше будет этот коэффициент. Учитывая этот факт, пересчитаем значения жилой площади. 

```python
# заменим 346 значение жилой площади, несоответствующее требованиям законодательства
# для чего используем функцию value_changer со следующими параметрами
# value_for_change = значение living_area, value_min = 14, value_max = total_area
# new_value = значение total_area * la_ta_ratio

def area_changer(row):
    '''
    Функция area_changer сравнивает значение value_for_change 
    со значением value_treshold, и если значение value_for_change больше,
    возвращает измененное значение value_treshold.  
        
    Если же значение value_for_change меньше value_treshold,
    то функция возвращает измененное значение value_treshold.
    '''
    value_for_change = row['living_area']
    value_treshold = row['total_area']
    if value_for_change <= value_treshold:
        if value_for_change < 14:
            if (value_treshold * 0.6) >= 14:
                return value_treshold * 0.6
            if (value_treshold * 0.7) >= 14:
                return value_treshold * 0.7
            if (value_treshold * 0.8) >= 14:
                return value_treshold * 0.8
            if (value_treshold * 0.9) >= 14:
                return value_treshold * 0.9
            return value_treshold 
        return value_for_change
    return value_treshold * 0.6
   
df['living_area'] = df.apply(area_changer, axis=1)


# проверим результаты

display(df[['living_area', 'total_area']].describe())
print('Число объектов недвижимости жилой площадью менее 14 кв.м.: ', 
      len(df.query('living_area < 14')))
```

```python
# построим гистограмму, для изучения последствий обработки

df['living_area'].plot.hist(
    bins=100, legend=True, figsize=(5, 4), xlim=(0,300), ylim=(0, 5500)
)
df['total_area'].plot.hist(
    bins=100, alpha=0.3, legend=True, figsize=(5, 4), xlim=(0,300), ylim=(0, 5500)
)
plt.show()
```

**Промежуточные выводы:**

Обработка дала свои результаты. Превышение жилой площади над общей ликвидирован. Распределение жилой площади выглядит лучше и практически повторяет контур общей площади.

- последний;

```python
# проверим столбцы 'floor' и 'floors_total'

print('Данные столбца \'floor\'')
print('Число пропусков:', df['floor'].isna().sum())
display(df['floor'].describe())

print()

print('Данные столбца \'floors_total\'')
print('Число пропусков:', df['floors_total'].isna().sum())
display(df['floors_total'].describe())
```

```python
# добавим столбец, содержащий группировку этажа квартиры: 
# варианты — первый, последний, другой

# напишем функцию

def floor_classifier(row):
    '''
    Функция floor_classifier принимает в параметры таблицу,
    и возвращает: 
     - классификацию 'первый', если в row['floor'] равен 1;
     - классификацию 'последний', если row['floor'] равен значению 
     row['floors_total']
     - классификацию 'другой', в остальных случаях.
    
    '''
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'

# применим функцию, для создания нового столбца

df['floor_class'] = df.apply(floor_classifier, axis=1)

# проверим результат 

df.sample(5, random_state=1)
```

```python
# сохраним срез для контроля

control_room = df.query('rooms == 0').sort_values(by='living_area')

# напишем функцию

def room_classifier(row):
    '''
    Функция room_classifier принимает в параметры таблицу.
    Если значение в колонке ['rooms'] равно 0, то функция заменяет
    его на подходящее количество комнат, расчитанное исходя из
    медианного значения комнат для квартир приблизительно подходящей площади.
    Границы площади определены исходя из исследования среза квартир с 0 комнат.
    '''
    room = row['rooms']
    total_area = row['total_area']
    if room == 0:
        if total_area < 20:
            return df.query('total_area < 20')['rooms'].median()
        if total_area < 30:
            return df.query('total_area < 30')['rooms'].median()
        if total_area < 40:
            return df.query('total_area < 40')['rooms'].median()
        if total_area < 50:
            return df.query('total_area < 50')['rooms'].median()
        if total_area < 60:
            return df.query('total_area < 60')['rooms'].median()
        if total_area < 70:
            return df.query('total_area < 70')['rooms'].median()
        return df.query('190 <total_area < 215')['rooms'].median()
    return room

df['rooms'] = df.apply(room_classifier, axis=1)
df['rooms'] = df['rooms'].astype('int')

# проверим результат

display(df.query('rooms == 0').sort_values(by='living_area'))
display(display(df['rooms'].describe()))
df.loc[control_room.index.tolist()]
```

Пиковые значения выглядят недостоверно. Есть вероятность ошибочного заполнения поля с учетом кладовок и гардеробных, которые не являются комнатами и не должны входить в этот парамтер. Для надлежащей замены определим среднюю площадь комнаты 
элитных жилых объектов (площадью свыше 100 метров), а затем применим этот коэффициент для замены значений. Это объекты большой стоимостью, поэтому простое удаление их из выборки может привести к искажению данных. При этом комнаты оказывают лишь опосредованное влияние на формирование цены, поэтому их усредненное значение не исказит данные.

Используем в расчетах жилую площадь, так как комнаты являются жилыми помещениями.

```python
# оценим количество объектов, с жилой площадью свыше 100 метров

print(len(df.query('living_area > 100')))

# посчитаем средний размер комнат для таких объектов 
# исходя из жилой площади, так как комнаты - это жилые помещения

rooms_k_for_elite = df.query('living_area > 100')['living_area'].mean() / df.query('living_area > 100')['rooms'].mean() 
display(df.query('living_area > 100')['living_area'].median() / df.query('living_area > 100')['rooms'].median())
rooms_k_for_elite
```

**Промежуточные выводы:**
В датасете содержится 395 объектов большой площадью. Это достаточное значение, чтобы положиться на использование среднего значения при расчете коэффициента площади на комнату. При этом небольшое количество крайних значений не должно сильно исказить значения. Коэффициент площади на комнату отличается на 0.49 при использовании медианы для расчета. Произведем замену пиковых значений согласно вычисленному коэффициенту. Используем уже определенную функцию room_classifier, чтобы доработать её и произвести замену значений.

```python
def room_classifier_for_elite(row):
    '''
    Функция room_classifier_for_elite принимает в параметры таблицу.
    Если значение в колонке ['rooms'] больше 9, то функция заменяет
    его на подходящее количество комнат, расчитанное исходя из
    коэффициента площади на комнату.
    '''
    room = row['rooms']
    total_area = row['living_area']
    if room > 9:
        return int(total_area / 26.69)
    return room

df['rooms'] = df.apply(room_classifier_for_elite, axis=1)
df['rooms'] = df['rooms'].astype('int')

# проверим результат

display(df.query('rooms > 9'))
display(display(df['rooms'].describe()))
```

Небольшой спад на количество трехкомнатных и дальнейшее последовательное уменьшение, связанное с увеличением комнат. Распределение Пуассона также подтверждает реальность. ### Влияние даты продажи на ценообразование

Уточним влияние даты продажи на ценовую политику рынка недвижимости, для этого сперва определим дату продажи.

```python
# добавим новый столбец 'дата продажи'
# для этого определим функцию 

def time_machine(row):
    '''
    Функция time_machine принимает в параметры таблицу
    и увеличивает дату публикации объявления из столбца 'publication_data'
    на число дней, указанных в столбце 'days_exposition',
    после чего возвращает обновленную дату.
    '''
    time_delta = row['days_exposition']
    old_data = row['publication_data']
    new_date = old_data + pd.Timedelta(days=time_delta)
    return new_date

# применим функцию и проверим результы

df['sale_date'] = df.apply(time_machine, axis=1)
df['sale_year'] = df['sale_date'].dt.year
df[['publication_data', 'sale_date', 'sale_year', 'days_exposition']].head()
```

Посчитаем среднюю стоимость проданных объектов недвижимости исходя из года в Санкт-Петербурге и Ленинградской области.

```python
# посчитаем среднюю стоимость объектов исходя из года для Ленобласти

lo_mean_per_year = (
    df.query('locality_name != "санкт-петербург"')
    .pivot_table(index='sale_year', 
     values='last_price', 
     aggfunc=('mean', 'min', 'max', 'median'))
)

# посчитаем среднюю стоимость объектов исходя из года для Санкт-Петербурга

spb_mean_per_year=(
    df.query('locality_name == "санкт-петербург"')
    .pivot_table(index='sale_year', 
     values='last_price', 
     aggfunc=('mean', 'min', 'max', 'median'))
)

# посчитаем среднюю стоимость объектов исходя из года для центра Санкт-Петербурга

spb_center_mean_per_year=(
    df.query('locality_name == "санкт-петербург" and km_to_center < 4')
    .pivot_table(index='sale_year', 
     values='last_price', 
     aggfunc=('mean', 'min', 'max', 'median'))
)

lo_mean_per_year = lo_mean_per_year.reset_index()
spb_mean_per_year = spb_mean_per_year.reset_index()
spb_center_mean_per_year = spb_center_mean_per_year.reset_index()

display('Изменение ценовых показателей в ЛО', lo_mean_per_year)
print()
display('Изменение ценовых показателей в Санкт-Петербурге', spb_mean_per_year)
print()
display('Изменение ценовых показателей в центре Санкт-Петербурга', spb_center_mean_per_year)
print()
print('Количество продаж в 2015 году: ', len(df.query('sale_year == 2015')))
print('Количество продаж в 2016 году: ', len(df.query('sale_year == 2016')))
print('Количество продаж в 2017 году: ', len(df.query('sale_year == 2017')))
print('Количество продаж в 2018 году: ', len(df.query('sale_year == 2018')))
print('Количество продаж в 2019 году: ', len(df.query('sale_year == 2019')))
```

На 2015 год приходится всего 53 продажи. В рамках всего массива, это абсолютно нерепрезентативно. Удалим из графика эту дату, чтобы не искажать данные.

Особенно, когда оно предоставляет дополнительные преференции, кроме безопасности или шумоизоляции, например, видовые. Поэтому там, где подобные преимущества более часты, их ценность выше. При этом, не стоит забывать о том, что для центра и области характерна малоэтажная застройка. А значит, количество квартир на первых и последних этажах банально больше.
