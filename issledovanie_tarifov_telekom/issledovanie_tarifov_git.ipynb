# Сравнительный анализ тарифных планов "Смарт" и "Ультра"

Необходимо проивести сравнительный анализ двух тарифных планов: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег. Исследование будет произведено на небольшой выборке клиентов (500 пользователей). Данные содержат: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

**Цель исследования:**

Из тарифных планов "Смарт" и "Ультра" определить тариф, приносящий больше прибыли компании.

**Предварительные гипотезы:**

В целях проведения исследования, заказчиком сформулированы две гипотезы:
1. Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
2. Средняя выручка пользователи из Москвы отличается от выручки пользователей из других регионов.

**Дополнительные задачи по обработке данных:**

Посчитать для каждого пользователя:
- количество сделанных звонков и израсходованных минут разговора по месяцам;
- количество отправленных сообщений по месяцам;
- объем израсходованного интернет-трафика по месяцам;
- помесячную выручку с каждого пользователя (вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножьте на значение из тарифного плана; прибавьте абонентскую плату, соответствующую тарифному плану).

**Ход исследования**

Данные представлены заказчиком. О качестве данных ничего не известно, кроме описания столбцов. Таким образом, исследование пройдет в четыре этапа:

### Выручка с пользователя в месяц

Для расчета выручки, сперва напишем функцию, возвращающую значение выручки помесячно, после чего применим ее к исследованным данным для формирования новых дата фреймов, содержащих возвращенные значения (то есть, выручку).

```python
# определим функцию

def cashier(data, shaft, sub_ultra, sub_smart, 
            cost_ultra, cost_smart, divider=1):
    '''
    Функция cashier принимает в параметрах следующие аргументы:
    data - обрабатываемый дата фрейм, который будет изменен;
    shaft - наименование обрабатываемого столбца в data;
    sub_ultra - вычитаемое, на которое будет уменьшены все значения, 
    содержащиеся в data[shaft], если столбец data['tariff'] == 'ultra';
    sub_smart - вычитаемое, на которое будет уменьшены все значения, 
    содержащиеся в data[shaft] в остальных случаях;
    cost_ultra - множитель результата вычитания и деления, 
    если столбец data['tariff'] == 'ultra';
    cost_smart - множитель результата вычитания и деления, в остальных случаях;
    divider - делитель, на который будет разделено уменьшенное значение data[shaft],
    значение по умолчанию 1.
    --
    Если разница data[shaft] и sub_ultra или sub_smart равно или меньше 0,
    то функция возвращает 0. В остальных случаях, функция cashier 
    пытается (try) возвратить округленный в большую сторону методом math.ceil
    результат, содержащийся в data[shaft], уменьшенный на sub_ultra 
    или sub_smart (смотри описание параметров) и разделенный на divider.
    После чего умножает на cost_ultra или cost_smart (смотри описание параметров). 
    Если данные содержат неокругляемые значения (NaN), то функция 
    повторяет (except) процедуру без округления.
    '''
    if data['tariff'] == 'ultra':
        if (data[shaft] - sub_ultra) <= 0:
            return 0
        try:
            return math.ceil((data[shaft] - sub_ultra) / divider) * cost_ultra 
        except:
            return (data[shaft] - sub_ultra) / divider * cost_ultra 
    else:
        if (data[shaft] - sub_smart) <= 0:
            return 0
        try:
            return math.ceil((data[shaft] - sub_smart) / divider) * cost_smart
        except:
            return (data[shaft] - sub_smart) / divider * cost_smart
```

##### Расчеты требуемых статистических значений

```python
# напишем функцию, которая вернет нам требуемые статистические значения

def statistical_oracle(data_for_stat, columns_for_stat=month_col):
    '''
    Функция statistical_oracle принимает массив данных в параметр data_for_stat,
    после чего перебирает все колонки, указанные в параметре columns_for_stat
    (значение по умолчанию columns_for_stat = month_col), 
    преобразовывает каждую колонку в формат nampy.array, 
    добавляя значение из нее в список ultra_values,
    если значение с таким же индексом из data_for_stat['tariff'] == 'ultra',
    либо в список smart_values, во всех остальных случаях.
    После чего для каждого списка возвращает значения стандартного отклонения,
    дисперсии и среднего, сопровождая вывод отформатированным print,    
    с выравниваем по левому краю, с фиксированной шириной строки 40 символов и
    округлением float до 4 знаков.
    '''
    ultra_values = []
    smart_values = []
    t_template = '{:<40}'
    n_template = '{:<40,.4f}'
    for i in columns_for_stat:
        for index, value in np.ndenumerate(np.array(data_for_stat[i]).flatten()):    
            if np.array(data_for_stat['tariff']).flatten()[index] == 'ultra':
                ultra_values.append(value)
            else:
                smart_values.append(value)
    return print('\n', '1. Для тарифа Ультра:', '\n', 
                 t_template.format('1.1. Стандартное отклонение выборки:'), 
                 n_template.format(np.nanstd(ultra_values, ddof=1)), '\n',
                 
                 t_template.format('1.2. Дисперсия выборки: '), 
                 n_template.format(np.nanvar(ultra_values, ddof=1)), '\n',
                 
                 t_template.format('1.3. Среднее значение выборки: '), 
                 n_template.format(np.nanmean(ultra_values)), '\n\n',
                 
                 '2. Для тарифа Смарт', '\n',
                 t_template.format('2.1. Стандартное отклонение выборки: '), 
                 n_template.format(np.nanstd(smart_values, ddof=1)), '\n',
                 
                 t_template.format('2.2. Дисперсия выборки: '), 
                 n_template.format(np.nanvar(smart_values, ddof=1)), '\n',
                 
                 t_template.format('2.3. Среднее значение выборки: '),
                 n_template.format(np.nanmean(smart_values)))
```

```python
# прогоним функцию циклом по нужным нам данным

stat_data_list = [messages_per_month, mb_per_month, min_per_month]

for i in stat_data_list:
    if len(i) == len(messages_per_month):
        print('\033[1m' + 'I. Статистические значения по отправленным сообщениям' + '\033[0m', end='')
        statistical_oracle(i)
        print()
    elif len(i) == len(mb_per_month):
        print('\033[1m' + 'II. Статистические значения по потраченным мегабайтам' + '\033[0m', end='')
        statistical_oracle(i)
        print()
    else:    
        print('\033[1m' + 'III. Статистические значения по потраченным минутам' + '\033[0m', end='')
        statistical_oracle(i)        
```

### Статистическая достоверность исследования

Предположим, что представленная выборка является репрезентативной. При учете того, что число наблюдений в выборке больше 30, можно использовать ее для проведения статистического исследования.

Рассмотрим предустановленные вопросы заказчика:

1. Средняя выручка пользователей тарифов Ультра и Смарт различаются;
2. Средняя выручка пользователи из Москвы отличается от выручки пользователей из других регионов.

Для их проверки, сформулируем гипотезы:

**1. По вопросу различия средней выручки пользователей тарифов Ультра и Смарт**
- Н0: Средняя выручка пользователей тарифов Ультра и Смарт равна.
- Н1: Средняя выручка пользователей тарифов Ультра и Смарт отличается.

**2. По вопросу различия средней выручки пользователей тарифов из Москвы и других городов**
- Н0: Средняя выручка пользователей тарифов из Москвы и других городов равна.
- Н1: Средняя выручка пользователей тарифов из Москвы и других городов отличается.

Так как у нас есть основания предполагать, что выборки были софрмированы случайным способом, можно считать их независимыми. Таким образом, для проверки статистической вероятности гипотез, проведем их сравнение методом ttest_ind

Изучим первый вопрос и проверим гипотезы.

**1. По вопросу различия средней выручки пользователей тарифов Ультра и Смарт**
- Н0: Средняя выручка пользователей тарифов Ультра и Смарт равна.
- Н1: Средняя выручка пользователей тарифов Ультра и Смарт отличается.

```python
sample_ultra = np.array(total_df_users.query('tariff == "ultra"')['total_income']).flatten()

sample_smart = np.array(total_df_users.query('tariff == "smart"')['total_income']).flatten()

alpha = 0.01

results = st.ttest_ind(sample_ultra, sample_smart)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Промежуточный вывод:**

Нулевая гипотеза отвергнута. Крайне низкое р-значение демонстрирует, что вероятность получить случайно результаты наших выборок, средние значения которых не равны, в то время как они были бы  равны, маловероятна. 

Следовательно, со значительной вероятностью, средняя выручка пользователей тарифов Ультра и Смарт отличается.

Изучим второй вопрос и проверим гипотезы

**2. По вопросу различия средней выручки пользователей тарифов из Москвы и других городов**
- Н0: Средняя выручка пользователей тарифов из Москвы и других городов равна.
- Н1: Средняя выручка пользователей тарифов из Москвы и других городов отличается.

```python
sample_moscow = np.array(total_df_users.query('city == "москва"')['total_income']).flatten()

sample_others = np.array(total_df_users.query('city != "москва"')['total_income']).flatten()

alpha = 0.01

results = st.ttest_ind(sample_moscow, sample_others)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```
