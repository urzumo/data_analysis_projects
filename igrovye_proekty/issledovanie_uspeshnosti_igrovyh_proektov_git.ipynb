# ИССЛЕДОВАНИЕ УСПЕШНОСТИ ИГРОВЫХ ПРОЕКТОВ

## Цели исследования

Необходимо провести исследование для интернет-магазина «Стримчик» (далее - заказчик), который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 

**Цель исследования:**

Выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании, для этого необходимо:

**Задачи исследования:**

1. Посмотрите, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?


2. Посмотрите, как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?


3. Возьмите данные за соответствующий актуальный период. Актуальный период определите самостоятельно в результате исследования предыдущих вопросов. Основной фактор — эти данные помогут построить прогноз на 2017 год.


4. Не учитывайте в работе данные за предыдущие годы.


5. Какие платформы лидируют по продажам, растут или падают? Выберите несколько потенциально прибыльных платформ.


6. Постройте график «ящик с усами» по глобальным продажам игр в разбивке по платформам. Опишите результат.


7. Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.


8. Соотнесите выводы с продажами игр на других платформах.


9. Посмотрите на общее распределение игр по жанрам. Что можно сказать о самых прибыльных жанрах? Выделяются ли жанры с высокими и низкими продажами?


10. Составить портрет пользователя каждого региона (NA, EU, JP), для чего необходимо определить:
 - Самые популярные платформы (топ-5). Опишите различия в долях продаж.
 - Самые популярные жанры (топ-5). Поясните разницу.
 - Влияет ли рейтинг ESRB на продажи в отдельном регионе?


11. Проверить предварительные гипотезы, предустановленные Заказчиком:
 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
 - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

**Дополнительные задачи при предобработке данных:**

1. Обратить внимание на аббревиатуру 'tbd' в столбцах с рейтингом. Отдельно разберите это значение и опишите, как его обработать;


2. Посчитайте суммарные продажи во всех регионах и запишите их в отдельный столбец.

**Особенности представленного датасета:**
1. Представлены данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.


2. В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».


**Ход исследования:**

Данные представлены заказчиком. О качестве данных ничего не известно, кроме описания столбцов. Таким образом, исследование пройдет в четыре этапа:
1. Обзор данных.
2. Предобработка данных.
3. Исследование гипотез.
    1. Дополнительные расчеты.
    2. Изучение предварительных гипотез.
    3. Постановка новых гипотез (при необходимости).
    4. Изучение новых гипотез (при необходимости).

### Статистическое исследование предустановленных гипотез

Предположим, что в рамках исследуемого датасета нам доступна вся историческая совокупность данных по продажам игровых тайтлов. Таким образом, считая исследуемые данные генеральной совокупностью, можно рассмотреть и сформулировать следующие гипотезы, исходя из предустановленных заказчиком.

Для их проверки, сформулируем гипотезы:

**1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**
- $Н_0$: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- $Н_1$: Средние пользовательские рейтинги платформ Xbox One и PC отличаются.

**2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные**
- $Н_0$: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые.
- $Н_1$: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») отличаются.

#### Сравнение пользовательских рейтингов платформ xbox one и pc

 Изучим первый вопрос

**1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые**
- $Н_0$: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- $Н_1$: Средние пользовательские рейтинги платформ Xbox One и PC отличаются.

```python
# проверим количество наблюдений, чтобы удостовериться в надежности исследования

sample_xone = np.array(filtered_df.query('platform == "xone"')['user_score'].dropna()).flatten()

sample_pc = np.array(filtered_df.query('platform == "pc"')['user_score'].dropna()).flatten()

display(len(sample_xone))
display(len(sample_pc))
```

```python
alpha = 0.05

results = st.ttest_ind(sample_xone, sample_pc)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Промежуточные выводы:**


Нулевую гипотезу не получилось отвергнуть. С вероятностью 14% можно получить одинаковые средние значения. Соответственно, статистически вероятной можно считать Н0. А именно: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

#### Сравнение пользовательских рейтингов жанров action и sports

Изучим второй впорос.

**2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные**
- Н0: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые.
- Н1: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») отличаются.

```python
# проверим количество наблюдений, чтобы удостовериться в надежности исследования

sample_action = np.array(filtered_df.query('genre == "action"')['user_score'].dropna()).flatten()

sample_sport = np.array(filtered_df.query('genre == "sports"')['user_score'].dropna()).flatten()

display(len(sample_action))
display(len(sample_sport))
```

```python
alpha = 0.05

results = st.ttest_ind(sample_action, sample_sport)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Промежуточный вывод:**


Нулевая гипотеза отвергнута. Следовательно, со значительной вероятностью, средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») отличаются.

Несмотря на высокую популярность (продажи, рейтинг) других платформ, количество релизов по ним, а также общий спад продаж, неминуемо свидетельствует об утере актуальности продуктов, расчитанных на эти платформы.


3. Если необходимо выбрать одну платформу для рекламной кампании, то однозначной рекомендацией становится ps4, как консоль нового поколения, набирающая популярность резкими темпами. Кроме того, интерес к ней представлен во всех регионах, что позволит покрыть максимальный процент потенциальных клиентов.


4. В безопасный топ-2 рекомендаций по платформам входит xone, обеспечивающий дополнительное покрытие в регионах na и ea.


5. При дополнительном ориентировании на японский рынок, стоит включить в рекламную кампанию 3ds (замыкает топ-5 популярности в регионах na и еа, уверенно лидирует в jp. 


6. При более глубокой ориентации на jp рынок, стоит изучить возможность рекламы psv и wiiu.


7. В рекламной кампании конкретного продукта, внутри регионов eu и na, стоит делать ставки на следующие качества рекламируемого продукта (в порядке убывания важности):
 - Жанр игры: для безопасной и консервативной кампании стоит сконцентрироваться на жанрах action и shooter, в случае дополнительных возможностей, в список можно добавить sports и role-playing;
 - Возрастной рейтинг М для большинства аудитории более предпочтителен;
 - Высокие оценки критиков - косвенно могут повлиять на продажи и привлечение новых клиентов;
 - Не стоит ориентироваться на пользовательские оценки, в продвижении продукта, так как они никаким образом не связаны с продажами.
 
 
 8. Для рекламной кампании внутри региона jp необходимо ориентироваться на следубщие качества (в порядке убывания важности):
  - Жанр игры: следует сделать ставку на role-playing и action. В исключительных случаях, можно попробовать рекламу жанра fighting.
  - Возрастной рейтинг Т и E. 
  - Несмотря на более лояльную аудоиторию в данном регионе, общие статистические закономерности не нарушаются. И пользовательские оценки не оказывают никакого влияния на продажи игр. Оценки критиков, в целом более сдержанные, в отношении даже удачных проектов. Но это можно считать проявлением японской культуры (учитывая тот факт, что 80 баллов - крайне высокая оценка для японских тайтлов).
 
 
9. Отдельно необходимо подчеркнуть распространившийся в последнее время "ревью-бомбинг", когда недовольные пользователи начинают массово занижать рейтинг продукта, независимо от его качества.
