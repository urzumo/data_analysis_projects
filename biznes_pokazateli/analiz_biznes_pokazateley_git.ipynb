#### Retention Rate

Рассчитаем Retention Rate, применяя когортный анализ. Покажем изменения метрики во времени на графике. Найдём средний 
Retention Rate на второй месяц «жизни» когорт. Для построения когорт определим событие и период. 

Так как приложение интересно с позиций продаж и рекламы (показов), сформируем два набора когорт - по активности и по заказам.

##### RR когорт по активности

Возьмём дату, когда пользователь впервые проявил активность на сайте.

```python
# создадим новую переменную, на основании таблицы visits 
# после чего дополним ее данными о первом визите пользователя

first_activity_date = visits.groupby('uid')['start_ts'].min()
first_activity_date.name = 'first_activity_date'
activity_df = visits.join(first_activity_date, on='uid').copy()
```

```python
# проверим результат

activity_df.head()
```

Для формирования месячных когорт, определим месяц первой активности на сайте.

```python
activity_df['first_activity_month'] = activity_df['first_activity_date'].astype(
    'datetime64[M]'
)

activity_df['activity_month'] = activity_df['start_ts'].astype('datetime64[M]')
```

```python
# проверим результат

activity_df.head()
```

```python
# сгруппируем когорты по месячной активности

activity_pivot = activity_df.pivot_table(
    index='first_activity_month',
    columns='activity_month',
    values='uid',
    aggfunc='nunique'
)
```

```python
# построим хитмап изменения месячной активности 

plt.figure(figsize=(13, 9))
plt.title('Количество активных пользователей в когорте')
sns.heatmap(
    activity_pivot, annot=True, fmt='.1f', linewidths=1, 
    linecolor='gray', vmin=50, vmax=2000,
    xticklabels=activity_pivot.columns.strftime('%Y-%m'),
    yticklabels=(activity_pivot.reset_index()['first_activity_month'].dt.strftime('%Y-%m')))
plt.yticks(rotation=0)
plt.show()
```

```python
# добавим в созданный фрейм лайфтайм когорты

activity_df['cohort_lifetime'] = (
    activity_df['activity_month'] - activity_df['first_activity_month']
)
activity_df['cohort_lifetime'] = activity_df[
    'cohort_lifetime'
] / np.timedelta64(1, 'M')

activity_df['cohort_lifetime'] = round(activity_df['cohort_lifetime']).astype('int')
```

```python
# проверим результат

activity_df.head()
```

```python
# сформируем когорты, с учетом лайфтайма

activity_cohorts = activity_df.groupby(
    ['first_activity_month','cohort_lifetime']).agg(
    {'uid':'nunique'}).reset_index()
```

```python
activity_cohorts
```

```python
# дополнима данные начальным количеством пользователей в каждой когорте
# для расчета RR

initial_users_count = activity_cohorts[activity_cohorts['cohort_lifetime'] == 0][
    ['first_activity_month', 'uid']
]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
```

```python
# дополним когорты полученными значениями

activity_cohorts = activity_cohorts.merge(initial_users_count,on='first_activity_month')
```

```python
# рассчитаем retetnion rate

activity_cohorts['retention'] = activity_cohorts['uid'] / activity_cohorts['cohort_users']
```

```python
# проверим результат

activity_cohorts.head()
```

```python
# построим сводную таблицу, содержащую RR

activity_retention_pivot = activity_cohorts.pivot_table(
    index='first_activity_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum'
)
```

```python
# проверим результат

activity_retention_pivot
```

```python
# построим хитмап изменения RR

sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(activity_retention_pivot, annot=True, fmt='.1%', linewidths=1, 
            linecolor='gray', vmin=0.01, vmax=0.1,
            xticklabels=activity_retention_pivot.columns,
            yticklabels=(activity_retention_pivot.reset_index()['first_activity_month'].dt.strftime('%Y-%m')))
plt.yticks(rotation=0);
```

```python
# посчитаем средний RR на 2 месяц жизни когорт

print('Средний retention rate на 2й месяц жизни когорт составляет:',
     round(activity_retention_pivot[1].mean() * 100, 2), '%.')
```

**Промежуточные выводы:**
1. После первого месяца происходит катастрофическое падение количества пользователей. Почти на 93,5%.
2. Падение RR после второго месяца крайне незначительное.
3. Можно сделать вывод, что пользователи, оставшиеся в приложении на 2 месяц, останутся в нем и дальше.
4. Следовательно, сайт хорошо справляется с удержанием постоянных пользователей. 
5. Для увеличения успешности сайта, необходимо изменить RR 2го месяца, а именно конверсию первичного захода в постоянного пользователя.

#### Среднее количество покупок на пользователя

Установим период, за который будем вычислять среднее количество заказов. В случае необходимости, эту переменную можно будет изменить. Возьмем период в 6 месяцев.

#### LTV

Выясним, как меняется LTV на покупателя по когортам. Учтем, что LTV — накопительная метрика. Рассчитаем средний LTV по когортам за 6 месяцев. В расчёт включим когорты, «прожившие» не менее 6 месяцев. Маржинальность сервиса — 100%. Отразим изменения метрики во времени на графике.

```python
# создадим датасет для вычисления LTV

ltv_df = orders_df.copy()
ltv_df['time_to_event'] = ltv_df['order_month'] - ltv_df['first_order_month']
```

```python
# посчитаем количество покупателей

cohort_sizes = (
    ltv_df.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']
```

```python
# проверим результат

cohort_sizes.head()
```

```python
ltv_cohorts = (
    ltv_df.groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
```

```python
# проверим результат

ltv_cohorts.sample(5, random_state=1)
```

```python
# дополним данными о количестве покупателей

report = pd.merge(cohort_sizes, ltv_cohorts, on='first_order_month')
```

```python
# проверим результат

report.head()
```

```python
# подготовим датасет для построения когорт

report['age'] = (
    report['order_month'] - report['first_order_month']
) / np.timedelta64(1, 'M')

report['age'] = report['age'].round().astype('int')
```

```python
# проверим результат

report.head()
```

```python
# найем ltv

report['ltv'] = report['revenue'] / report['n_buyers']
report.head()
```

```python
output = report.pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
).round(2)

output.fillna('')
```

```python
# отфильтруем когорты со сроком жизни 6 и более месяцев

filtered_output = output[(output[5] == output[5])]

# проверим результат

filtered_output
```

```python
# отфильтруем данные за 6 месяцев и проссумируем их

filtered_output.loc[:, 0: 5].sum(axis=1)
```

```python
# визуализируем 

plt.plot(filtered_output.loc[:, 0: 5].sum(axis=1), color='red')

plt.title('Изменение среднего LTV с 6 по 12 когорту', fontsize=14)
plt.xlabel('Когорта')
plt.ylabel('LTV')
plt.grid()
```
