```python
# предположим, что можно обеспечить более точное заполнение пропусков
# в столбце 'days_employed', если выявим связь с другими значениями
#  для этого посчитаем коэффициент корреляции между данными, 
# содержащимися в столбцах 'days_employed' и 'dob_years'

df['days_employed'].corr(df['dob_years'], "pearson")
```

Не лучшее значение, чтобы увеличить его, попробуем добавить новую классификацию по возрасту, выделив различные возрастные группы, после чего посчитать r уже на обновленных данных.
Сперва нужно определиться с возрастной периодизацией. В современной науке этот параметр распространен достаточно широко. Однако, с учетом социально-экономических особенностей устройства российского общества, использовать мировые классификации или классификацию ВОЗ нам представляется неразумным. К сожалению, в настоящее время, в российской науке отсутствует единая стандартизация в вопросах возрастной периодизации, поэтому мы сошлемся на следующее, до сих пор популярное, решение.


**Возрастная периодизаций 1965 года на симпозиуме Академии педагогических наук СССР:**

Данная периодизация представляется нам наиболее подходящей для представленной задачи, учитывая постсоветское наследие на территории Российской Федерации.

    - Юношеский возраст — от 17 до 21 года (муж.); от 16 до 20 лет (жен.)
    - Средний возраст
    - первый период — от 22 до 35 года (муж.); от 21 до 35 лет (жен.)
    - второй период — от 36 до 60 года (муж.); от 36 до 55 лет (жен.)
    - Пожилые люди — от 61 до 75 года (муж.); от 56 до 75 лет (жен.)

Объединим представленные выше группы в возрастные классы, разделив категорию 36-60 на две, соответствующих классификации ВОЗ:


- 1: до 21 года
- 2: 22-35 лет
- 3: 36-45 лет 
- 4: 46-60 лет
- 5: 61+

Напишем подходящую функцию, после чего добавим полученные данные в таблицу и пересчитаем r.

```python
# напишем функцию
def age_classifier(dob_years, age_class_1=21, 
                   age_class_2=35, age_class_3=45, age_class_4=60):
    '''
    Функция age_classifier принимает в параметры 'dob_years'
    возраст заемщика, после чего возвращает 
    соответствующее ему значение возрастной периодизации, в зависимости 
    от величины приданной функции сумме в параметрах age_class_*. 
    
    В качестве значения по умолчанию параметров age_class_*,
    используется выбранная нами модель возрастной периодизации,
    основанная на симпозиуме АПН 1965 года.
    
    '''
    if dob_years <= age_class_1:
        return 1
    if dob_years <= age_class_2:
        return 2
    if dob_years <= age_class_3:
        return 3
    if dob_years <= age_class_4:
        return 4
    return 5

# протестируем функцию для каждой группы
print(age_classifier(19))
print(age_classifier(27))
print(age_classifier(45))
print(age_classifier(55))
print(age_classifier(67))
```

```python
# добавим столбец с возрастной периодизацией в таблицу

df['age_class'] = df['dob_years'].apply(age_classifier)

# проверим результат работы функции

display(df.head())

# оценим получившийся разброс
df['age_class'].value_counts()
```

Отличное статистическое распределение, за исключением первой группы. Но её выделение обосновано экономически и законодательно, поэтому оставим ее в составе данных. Перейдем к рассчету коэффициента корреляции.

```python
df['days_employed'].corr(df['age_class'], "pearson")
```

Коэффициент уменьшился на 4%. Можно сделать вывод, что возрастная периодизация не имеет связи с трудовым стажем в данных, что безусловно связано с ошибкой заполнения столбца 'days_employed'. Сгруппируем среднее значение стажа по возрастным группам.

```python
# создадим сводные таблицы, вызвав функции для изучения цифр
display(df.pivot_table(index=['age_class'], values='days_employed', aggfunc=['max', 'min','mean', 'median']))
```

А вот и ошибка в данных. Данные искажены из-за крайних значений. При этом, максимальные значения для 4-5 возрастной группы выглядят легитимно, если переводить из дней в часы. И абсолютно чуждо для групп с 1 по 3. 


Максимальное значение, если оценивать его в часах будет равняться приблизительно 45 годам. Что похоже на усредненное значение по заемщикам. Косвенно этот факт подтверждается тем, что 4 возрастная группа (46-60 лет - возрастное попадание в стаж пракитчески идеально!), самая широкопредставленная в возрастном массиве. 


Из этого можно сделать вывод, что данные по стажу первично содержали большое количество пропусков, а после обработки, пропуски были заполнены средним значением по всем заемщикам. При этом, по какой-то причине, замена проихводилась в единицах (дни стажа * 24). Что и привело к искажению среднего стажа всех возрасных групп. К тому же нельзя забывать об очевидной ошибке в виде отрицательного значения стажа.

Логично предположить, что замену стоит произвести по прогнозируемому среднему стажу в возрастной группе, исходя из занятости с 18 лет, при этом нужно не обрабатывать пропуски, а провести сквозную полную замену из-за крайне низкого качества данных.

```python
# напишем функцию, рассчитывающую средний стаж, исходя из возрастной группы
# и законодательного возраста начала рабочих отношений 18 лет

def days_employed_oracle(dob_years, age_class_1=21, 
                   age_class_2=35, age_class_3=45, age_class_4=60):
    '''
    Функция days_employed_oracle принимает в параметры 'dob_years'
    возраст заемщика, после чего возвращает 
    соответствующее ему значение среднего трудового стажа
    от величины приданной функции сумме в параметрах age_class_*. 
    
    В качестве значения по умолчанию параметров age_class_*,
    используется выбранная нами модель возрастной периодизации,
    основанная на симпозиуме АПН 1965 года.
    
    Вычитание 18 обеспечивает отсутствие учета первых 18 лет
    жизни в качестве рабочего стажа.
    Умножение на 365 позволяет привести значение к заданным
    технической документацией параметрам (измерение стажа в днях).
    
    '''
    if dob_years <= age_class_1:
        return (age_class_1 - 18) / 2 * 365
    if dob_years <= age_class_2:
        return (age_class_2 + age_class_1 - 18) / 2 * 365
    if dob_years <= age_class_3:
        return (age_class_3 + age_class_2 - 18) / 2 * 365
    if dob_years <= age_class_4:
        return (age_class_4 + age_class_3 - 18) / 2 * 365
    return (65 + age_class_4 - 18) / 2 *365

# протестируем функцию для каждой группы
print(days_employed_oracle(19))
print(days_employed_oracle(27))
print(days_employed_oracle(45))
print(days_employed_oracle(55))
print(days_employed_oracle(67))
```

```python
# заменим пропуски в колонке 'days_employed' на среднее значение
# трудового стажа для возрастных групп

df['days_employed'] = df['dob_years'].apply(days_employed_oracle).astype('int')

# проверим результат работы, запросив количество пропусков, 
# а также изменение среднего значения столбца

mean_days_employed_before = df_control['days_employed'].abs().mean()
mean_days_employed_after = df['days_employed'].mean()

print('Количество пропусков в столбце \'days_employed\':', df['days_employed'].isna().sum())
print('Среднее значение трудового стажа в столбце \'days_employed\' до замены: ' 
      + str(round(mean_days_employed_before / 365))  + ' года.')
print('Среднее значение трудового стажа в столбце \'days_employed\' после замены: ' 
      + str(round(mean_days_employed_after / 365)) + ' года.')
print(f'Среднее значение столбца \'days_employed\' изменилось ' 
      f'на {mean_days_employed_before / mean_days_employed_after:.1%}')

df.sample(5, random_state=1)
```

16 групп. Уже лучше, но все еще слишком высокая детализация. Предложим следующую классификацию, основываясь на нелинейном росте зарплат:
1. Класс 0: 1/2 и менее медианной зарплаты по данным.
2. Класс 1: 2/3 и менее медианной зарплаты по данным.
3. Класс 2: медианная зарплата по данным и менее.
4. Класс 3: полторы и менее медианных зарплаты по данным. 
5. Класс 4: две и менее медианных зарплат по данным.
6. Класс 5: свыше двух медианных зарплат по данным.

Проведем классификацию, после чего оценим получившиеся объемы данных.

```python
# напишем функцию
def income_classifier(total_income, base=median_total_income_after):
    '''
    Функция income_classifier принимает в параметры 'total_income'
    числовое значение дохода, после чего возвращает 
    соответствующее ему значение классификации, в зависимости 
    от величины приданной функции сумме в параметре base. 
    
    В качестве значения по умолчанию параметра base
    выбрана переменная median_total_income_after,
    содержащая данные о медианном значении df['total_income']
    после заполнения пропусков.
    '''
    if total_income <= base / 2:
        return 0
    if total_income <= base / 3 * 2:
        return 1
    if total_income <= base:
        return 2
    if total_income <= base * 1.5:
        return 3
    if total_income <= base * 2:
        return 4
    return 5

# протестируем функцию для каждой группы
print(income_classifier(50000))
print(income_classifier(90000))
print(income_classifier(140000))
print(income_classifier(170000))
print(income_classifier(300000))
print(income_classifier(500000))
```

```python
# добавим столбец с категоризацией в таблицу
# для чего введем тестовую переменную, чтобы провести эксперименты

df_value = df
df_value['income_class'] = df_value['total_income'].apply(income_classifier)

# проверим результат работы функции

display(df_value.head())

# оценим получившийся разброс
df_value['income_class'].value_counts()
```

Это вполне может стать интересным дополнительным кейсом, который окажет влияние на общие сведения. Но в настоящее время, это субисследование не требуется. **Вывод:** значит можно объединить все сделки в "получение образования".


5. Кредиты на проведение свадьбы не требуют дополнительных пояснений и данные по ним могут быть использованы as is.

**Итоги:** выделим следующие ключевые слова "недвижимость", "автомобиль", "образование", "жилье", "свадьба". Проведем стемминг, для выяснения основы слова (чтобы осуществить замену в таблице), а также заменим все сущности "жилья" на "недвижимость.

Перейдем к обработке столбца 'purpose'.

```python
# напишем функцию, для возврата основы слова из 
# списка значений в целевом столбце

def stemmator(column, base=stem_credit_purpose, exception='жил', exception_return='недвижим'):
    '''
    Функция stemmator проверяет значения в строке 
    столбца, выбранного параметром column. 
    Если это значение есть в списке параметра base, 
    то фукнция возвращает это значение из списка параметра base.
    
    Функции можно придать дополнительные параметры:
    exception - исключение из параметра base, позволяющее 
    возвратить уникальное значение исключения, зафиксированное 
    в параметре exception_return.
    
    Текущие значения по умолчанию параметров exception и exception_return
    обусловлены необходимостью объединить сущности 'жил' и 'недвижим'
    
    '''
    for row in column:
        for word in base:
            if word == exception:
                if word in row:
                    return exception_return
            if word in row:
                return word

# обновим данные столбца 'purpose', разбив значения,
# чтобы функция обработки могла пройтись по всем значениям в строке
# после чего применим функцию

df['purpose_class'] = df['purpose'].str.split().apply(stemmator)

# проверим результат работы функции

df.sample(10, random_state=1)
```

**Промежуточные выводы:**

## Выводы

Здесь и далее, мы будем использовать термин ВКП - Вероятность возврата Кредита без Просрочки. Базовый ВКП - ВКП, рассчитанный на основании общих данных, представленных заказчиком, без учета влияния каких-либо характеристик заемщика.

ВКП без просрочки равняется 8.8%. У нас отсутствует информация о порядке формирования данного значения. Вполне вероятно, что в эту цифру попадают даже те, у кого есть просрочка в один-два дня. В связи с чем   сложно сделать вывод о плачевности ситуации. Однако, именно этот факт необходимо держать в голове при проведении анализа и оценке его итогов. 


По результатам проведенного исследования, можно прийти к выводу, что ни одна из гипотез, заявленных к анализу заказчиком, не имеет хоть сколь сильного влияния на ВКП. Представленные категории могут повлиять на цель кредита, его размер, но не на итоговую просрочку. 


Более того, представленные данные больше подходят для создания модели, предсказывающей вероятность ВКП у каждого конкретного клиента. Особенно, в случае уточнения термина "просрочка" и разбитие его на временные периоды, исходя из специфики банковской деятельности. Кроме того, полученная в ходе исследования информация, помогла выявить несовершенство в формировании представленных данных, о чем, безусловно, стоит информировать заказчика.

Учитывая достаточно однородную структуру данных, мы можем посчитать усредненный процент влияния категорий, озвученных заказчиком, на возврат кредита в срок.

```python
# напишем функцию, для подсчета значений, 
# заносимых в финальную таблицу

def totalizer(list_name, name, table):
    '''
    Функция totalizer служит для наполнения списков вычислениями.
    Она принимает в качестве параметра название списка (list_name),
    значение (name), которое добавляет в начало списка,
    таблицу (table) из которой берутся данные для вычисления.
    Функция возвращает значение (name), max() для базового и измененного, 
    min() для базового и измененного и mean() для измененного значений ВКП,
    создавая список из 6 элементов.
    '''
    list_name.append(name)
    list_name.append(table['debt'].max())
    list_name.append(table['debt'].min())
    list_name.append(table['total_impact'].max())
    list_name.append(table['total_impact'].min())
    list_name.append(table['total_impact'].mean())

# создадим списки, которые наполним функцией

children_total_impact_list = []
family_total_impact_list = []
income_total_impact_list = []
purpose_total_impact_list = []

# запустим функцию последовательно для каждого списка циклом

totalizer(children_total_impact_list, 'children', df_children_pivot)
totalizer(family_total_impact_list, 'family', df_family_pivot)
totalizer(income_total_impact_list, 'income', df_income_pivot)
totalizer(purpose_total_impact_list, 'purpose_class', df_purpose_pivot)

# создадим список списков, чтобы наполнить таблицу
# выведем его на экран, чтобы проверить работу функции

total_impact_list = [
    children_total_impact_list, family_total_impact_list,
    income_total_impact_list, purpose_total_impact_list
]

print(total_impact_list)

# соберем полученные данные в единую таблицу

impact_columns = [
    'impact_object', 'max_impact_value', 'min_impact_value', 
    'max_impact_delta', 'min_impact_delta', 'mean_impact_delta'
]

results = pd.DataFrame(data=total_impact_list, columns=impact_columns) 
results
```

**Рекомендации:** одинокий (но не разведенные) и состоящие в гражданском браке представляют приоритетную категорию заемщиков (при этом ВКП кредита на свадьбу для состоящих в гражданском браке демонстрирует самый худший ВКП 7,9% по сравнению с ближайшим 9,1% при сделках с недвижимостью). Вдоцвы и разведенные составляют худшую категорию заемщиков, имеющих стойкую тенденцию к снижению ВКП.


На последнем месте по изменению ВКП находится **цель кредита**. Кредиты на свадьбу и недвижимость имеют тенденцию к росту ВКП. **Рекомендации:** кредиты на авто и образование тяготеют к более высокому ВКП. Кредиты на свадьбу и недвижимость, зачастую связаны с большими тратами и склонны к снижению ВКП.
